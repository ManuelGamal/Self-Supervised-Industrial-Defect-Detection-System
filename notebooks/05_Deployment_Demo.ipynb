# Deployment Demo

This notebook demonstrates the deployment API across all 6 categories. Make sure the API is running at `http://localhost:8000` via `docker-compose up`.

```python
import requests
import glob
from pathlib import Path

categories = ["bottle", "capsule", "carpet", "hazelnut", "leather", "pill"]
base_url = "http://localhost:8000"
data_dir = Path("C:/Users/manue/Downloads/archive (2)")

print("Healthcheck:", requests.get(f"{base_url}/health").json())
```

```python
for cat in categories:
    cat_dir = data_dir / cat
    # Get one image for demonstration
    images = list(cat_dir.rglob("*.png")) + list(cat_dir.rglob("*.jpg"))
    if not images:
        print(f"No image found for {cat}")
        continue
    
    img_path = images[0]
    with open(img_path, "rb") as f:
        response = requests.post(
            f"{base_url}/predict",
            data={"category": cat},
            files={"file": f}
        )
    
    print(f"--- {cat.upper()} ---")
    if response.status_code == 200:
        print("Result:", response.json())
    else:
        print("Error:", response.text)
```
